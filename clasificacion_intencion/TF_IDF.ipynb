se caragn los datos limpios y se checan los registros de cada categoría pra detectar desbalances, dicho desbalance detectado fue en la categoría de queja por lo que se optó un proceso diferente a dicha categoría.

```python
import pandas as pd

df = pd.read_csv("/content/buzon_limpio.csv")
df["etiqueta"].value_counts()
```

por lo mencionado anteriormente se opta por añadir ejemplos sintéticos para mejorar el aprendizaje del modelo en esta clase

```python
nuevas_quejas = [
    "El sistema presenta fallas constantes",
    "La atención fue muy lenta",
    "No resolvieron mi problema",
    "El trámite tardó demasiado",
    "La plataforma no funciona correctamente",
    "No recibí respuesta a mi solicitud",
    "El servicio fue deficiente",
    "Hubo errores durante el proceso",
    "El personal no dio seguimiento",
    "Estoy inconforme con la atención",
    "El sistema se cae frecuentemente",
    "La atención no fue adecuada",
    "No se solucionó el inconveniente",
    "La respuesta fue tardía",
    "El servicio no cumplió lo esperado"
]

df_quejas = pd.DataFrame({
    "comentario": nuevas_quejas,
    "etiqueta": "queja"
})

df = pd.concat([df, df_quejas], ignore_index=True)
df = df.sample(frac=1, random_state=42)

df["etiqueta"].value_counts()
```

se confih¿gura el TfidfVectorizer para convertir el texto en una matriz numérica. Se usan n-gramas de 1 a 3 palabras para capturar combinaciones de términos frecuentes.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib

vectorizador = TfidfVectorizer(
    ngram_range=(1, 3),
    min_df=2,
    max_df=0.85,
    sublinear_tf=True
)

X = vectorizador.fit_transform(df["comentario"])
```

```python
import os

# Create the directory for models if it doesn't exist
models_dir = "../models"
os.makedirs(models_dir, exist_ok=True)
joblib.dump(vectorizador, os.path.join(models_dir, "tfidf_vectorizer.pkl"))

# Create the directory for data if it doesn't exist
data_dir = "../data"
os.makedirs(data_dir, exist_ok=True)
df.to_csv(os.path.join(data_dir, "buzon_balanceado.csv"), index=False, encoding="utf-8")
```
